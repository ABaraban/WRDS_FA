```python
import wrds
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns

# Establish a connection to the WRDS database
db = wrds.Connection()

# Load the data
df = db.raw_sql("""
    SELECT
        i.dcn,
        i.seqnum,
        i.sectitle,
        i.tprice,
        i.tprice_adj,
        i.tprice_ar,
        i.shares,
        i.shares_adj,
        i.sigdate,
        i.secdate,
        i.ticker,
        i.acqdisp,
        i.optionsell,
        i.cleanse,
        c.prccd,
        c.datadate,
        c_next.prccd AS next_day_price,
        c.cusip
    FROM
        tr_insiders.table1 AS i
    LEFT JOIN
        comp_na_daily_all.secd AS c
    ON 
        i.cusip6 = LEFT(c.cusip, 6) AND
        c.datadate = i.secdate
    LEFT JOIN
        comp_na_daily_all.secd AS c_next
    ON
        i.cusip6 = LEFT(c_next.cusip, 6) AND
        c_next.datadate = i.secdate + INTERVAL '1 day'
    WHERE
        i.formtype = '4' AND
        i.acqdisp IS NOT NULL AND
        i.trancode IS NOT NULL AND
        i.cleanse IN ('R', 'H') AND
        ((i.trancode = 'P' AND i.acqdisp = 'A') OR (i.trancode = 'S' AND i.acqdisp = 'D'))
        LIMIT 1000
""")

# Print the number of rows before removing duplicates
print(f"Number of rows before removing duplicates: {len(df)}")

# Drop duplicate rows based on all columns to clean the DataFrame
df_cleaned = df.drop_duplicates()

# Print the number of rows after removing duplicates
print(f"Number of rows after removing duplicates: {len(df_cleaned)}")

# Check if `dcn` and `seqnum` uniquely identify rows in the initial data
if df[['dcn', 'seqnum']].duplicated().any():
    print("Initial data has duplicates based only on `dcn` and `seqnum`.")
    duplicates_initial = df[df.duplicated(subset=['dcn', 'seqnum'], keep=False)]
    print(duplicates_initial)
else:
    print("Initial data is uniquely identified by `dcn` and `seqnum`.")

# Check for duplicates based on the specified columns
columns_to_check = [
    'tprice', 'tprice_adj', 'tprice_ar', 'shares', 'sigdate',
    'secdate', 'ticker', 'acqdisp', 'optionsell', 'cleanse', 'seqnum', 'dcn'
]

# Find duplicates in the cleaned DataFrame
duplicates = df_cleaned[df_cleaned.duplicated(subset=columns_to_check, keep=False)]

# Print results and diagnostics
if not duplicates.empty:
    print("The following rows are not uniquely identified by the specified columns:")
    print(duplicates)
else:
    print("The specified columns uniquely identify each row.")
```

```python
df['secdate'] = pd.to_datetime(df['secdate'])

# Calculate percentage gain
df['profit'] = (df['prccd'] - df['tprice']) * df['shares']

# Extract year from secdate
df['year'] = df['secdate'].dt.year

# Filter purchase and sale transactions respectively
purchase_df = df[df['acqdisp'] == 'A']
sale_df = df[df['acqdisp'] == 'D']

# Group by year and calculate the average percentage gain
purchase_avg_gain = purchase_df.groupby('year')['profit'].mean().reset_index()
sale_avg_gain = sale_df.groupby('year')['profit'].mean().reset_index()

# Plotting
plt.figure(figsize=(12, 6))
sns.lineplot(data=purchase_avg_gain, x='year', y='profit', label='Average Purchase Percentage Gain')
sns.lineplot(data=sale_avg_gain, x='year', y='profit', label='Average Sale Percentage Gain')
plt.title('Average Percentage Gain by Year for Purchases and Sales')
plt.xlabel('Year')
plt.ylabel('Average Percentage Gain (%)')
plt.legend()
plt.grid(False)
plt.show()
```

```python
import wrds
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns

# Establish a connection to the WRDS database
db = wrds.Connection()

# Fetch and merge the data
df = db.raw_sql("""
    SELECT
        i.dcn,
        i.seqnum,
        i.sectitle,
        i.tprice,
        i.tprice_adj,
        i.tprice_ar,
        i.shares,
        i.shares_adj,
        i.sigdate,
        i.secdate,
        i.ticker,
        i.acqdisp,
        i.optionsell,
        i.cleanse,
        i.cusip6,
        c.prccd,
        c.datadate,
        c.cusip
    FROM
        tr_insiders.table1 AS i
    LEFT JOIN
        comp_na_daily_all.secd AS c
    ON 
        i.cusip6 = LEFT(c.cusip, 6) AND
        c.datadate = i.secdate
    WHERE
        i.formtype = '4' AND
        i.acqdisp IS NOT NULL AND
        i.trancode IS NOT NULL AND
        i.cleanse IN ('R', 'H') AND
        ((i.trancode = 'P' AND i.acqdisp = 'A') OR (i.trancode = 'S' AND i.acqdisp = 'D'))
        LIMIT 10000
""")

# Print initial row count
print(f"Number of rows before removing duplicates: {len(df)}")

# Check for duplicates post-merge
duplicates = df[df.duplicated(subset=['dcn', 'seqnum', 'ticker', 'acqdisp', 'cusip', 'datadate'], keep=False)]

if not duplicates.empty:
    print("Duplicates found post-merge:")
    print(duplicates)

    # Optional: Review a small number of duplicates
    print("Example duplicates:")
    print(duplicates.head())
else:
    print("No duplicates found post-merge.")

# Drop duplicate rows based on all columns
df_cleaned = df.drop_duplicates(subset=['dcn', 'seqnum'])

# Verify final number of unique rows
print(f"Number of rows after removing duplicates: {len(df_cleaned)}")

### Analysis: Average Percentage Gain By Year for Purchases and Sales

# Convert secdate to datetime
df_cleaned['secdate'] = pd.to_datetime(df_cleaned['secdate'])

# Calculate percentage gain
df_cleaned['percentage_gain'] = (df_cleaned['next_day_price'] - df_cleaned['tprice']) / df_cleaned['tprice'] * 100

# Extract year from secdate
df_cleaned['year'] = df_cleaned['secdate'].dt.year

# Filter purchase and sale transactions respectively
purchase_df = df_cleaned[df_cleaned['acqdisp'] == 'A']
sale_df = df_cleaned[df_cleaned['acqdisp'] == 'D']

# Group by year and calculate the average percentage gain
purchase_avg_gain = purchase_df.groupby('year')['percentage_gain'].mean().reset_index()
sale_avg_gain = sale_df.groupby('year')['percentage_gain'].mean().reset_index()

# Plot the results
plt.figure(figsize=(12, 6))
sns.lineplot(data=purchase_avg_gain, x='year', y='percentage_gain', label='Average Purchase Percentage Gain')
sns.lineplot(data=sale_avg_gain, x='year', y='percentage_gain', label='Average Sale Percentage Gain')
plt.title('Average Percentage Gain by Year for Purchases and Sales')
plt.xlabel('Year')
plt.ylabel('Average Percentage Gain (%)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import wrds
import pandas as pd

# Establish a connection to the WRDS database
db = wrds.Connection()

# Fetch insider data
insiders_df = db.raw_sql("""
    SELECT
        dcn,
        seqnum,
        sectitle,
        tprice,
        tprice_adj,
        tprice_ar,
        shares,
        shares_adj,
        sigdate,
        secdate,
        ticker,
        acqdisp,
        optionsell,
        cleanse,
        cusip6
    FROM
        tr_insiders.table1
    WHERE
        formtype = '4' AND
        acqdisp IS NOT NULL AND
        trancode IS NOT NULL AND
        cleanse IN ('R', 'H') AND
        ((trancode = 'P' AND acqdisp = 'A') OR (trancode = 'S' AND acqdisp = 'D'))
""")

# Fetch Compustat data
compustat_df = db.raw_sql("""
    SELECT
        cusip,
        datadate,
        prccd
    FROM
        comp_na_daily_all.secd
    LIMIT 10000000 -- Limit for practical purposes; adjust as necessary
""")

# Merge the datasets on cusip6 and the first 6 digits of cusip and datadate
merged_df = pd.merge(
    insiders_df,
    compustat_df,
    left_on=['cusip6', 'secdate'],
    right_on=[compustat_df['cusip'].str[:6], 'datadate'],
    suffixes=('_insider', '_compustat')
)

# Print the result to inspect
print(f"Total rows in the merged dataset: {len(merged_df)}")
print(merged_df.head())

# Identify all possible matches where datadate and cusip have multiple prccd values
duplicates = merged_df.groupby(['cusip', 'datadate']).filter(lambda x: x['prccd'].nunique() > 1)

if not duplicates.empty:
    print("Conflicting prccds found (same cusip and datadate but different prccd):")
    print(duplicates)
else:
    print("No conflicting prccds found.")

# Save the merged DataFrame to a file for further inspection
merged_df.to_csv('merged_data_cusip_date.csv', index=False)
```

```python
# Check for uniqueness of `compustat_df` on `cusip` and `datadate`
import wrds
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns

# Establish a connection to the WRDS database
db = wrds.Connection()
compustat_df = db.raw_sql("""
    SELECT *
    FROM
        comp_na_daily_all.secd
    LIMIT 100000 -- Limit for practical purposes; adjust as necessary
""")

duplicates = compustat_df[compustat_df.duplicated(subset=['cusip', 'datadate'], keep=False)]

if not duplicates.empty:
    print("Compustat data has duplicates based on `cusip` and `datadate`:")
    print(duplicates)
    print(f"Number of duplicate rows: {len(duplicates)}")
else:
    print("Compustat data is unique on `cusip` and `datadate`.")

# Optional: Summary of duplicates for further inspection
duplicate_summary = duplicates.groupby(['cusip', 'datadate']).size().reset_index(name='count')
print(duplicate_summary)
```
